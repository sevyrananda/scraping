```python
import pandas as pd # Digunakan untuk mengimpor modul Pandas ke dalam script.
import matplotlib.pyplot as plt # Digunakan untuk mengimpor modul Matplotlib ke dalam script.

# Digunakan untuk membaca file CSV dengan nama "data_aksesoris.csv".
data = pd.read_csv("data_product.csv")

# Digunakan untuk membuat sebuah figur baru dalam visualisasi dengan ukuran 12x7 inci.
plt.figure(figsize=(12, 7))

data['Address'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)  # Membuat pie chart dengan persen.
plt.ylabel('') # Menghapus label sumbu Y.

# Menambahkan judul.
plt.title("Visualisasi Produk Peralatan Makan dan Minum Berdasarkan Alamat menggunakan Pie Chart")

# Menyimpan Pie Chart dalam format JPG dengan nama file 'PieChart_Aksesoris.jpg'.
plt.savefig('img/PieChart_product.jpg', dpi=300, bbox_inches='tight')

# Menampilkan hasil Pie Chart.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV dengan nama "data_product.csv".
data = pd.read_csv("data_product.csv")

# Membuat sebuah figur baru dalam visualisasi dengan ukuran 12x7 inci.
plt.figure(figsize=(12, 7))

# Menghitung jumlah kemunculan tiap nilai dalam kolom 'Rating'.
data['Rating'].value_counts().sort_index().plot(kind='bar')

# Menambahkan judul dan label sumbu.
plt.title("Distribusi Rating Produk")
plt.xlabel("Rating")
plt.ylabel("Jumlah Produk")

# Menampilkan hasil grafik.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV dengan nama "data_product.csv".
data = pd.read_csv("data_product.csv")

# Menghitung jumlah kemunculan tiap nilai dalam kolom 'Rating'.
rating_counts = data['Rating'].value_counts()

# Membuat sebuah figur baru dalam visualisasi dengan ukuran 7x7 inci.
plt.figure(figsize=(7, 7))

# Membuat grafik pie.
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)

# Menambahkan judul.
plt.title("Proporsi Rating Produk")

# Menampilkan hasil grafik.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV dengan nama "data_product.csv".
data = pd.read_csv("data_product.csv")

# Menghitung frekuensi kemunculan tiap produk.
product_frequency = data['Name'].value_counts()

# Mengambil 10 produk dengan frekuensi kemunculan tertinggi.
top_products = product_frequency.head(10)

# Membuat sebuah figur baru dalam visualisasi dengan ukuran 12x7 inci.
plt.figure(figsize=(12, 7))

# Membuat grafik bar untuk visualisasi produk yang paling laris.
top_products.plot(kind='bar')

# Menambahkan judul dan label sumbu.
plt.title("10 Produk Paling Laris")
plt.xlabel("Produk")
plt.ylabel("Frekuensi Kemunculan")

# Mengatur label sumbu x agar tidak saling tumpang tindih.
plt.xticks(rotation=45, ha="right")

# Menampilkan hasil grafik.
plt.tight_layout()
plt.show()
```
